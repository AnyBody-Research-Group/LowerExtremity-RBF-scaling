```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.size']=8
mpl.rcParams['font.sans-serif']='Arial'
mpl.rcParams['figure.dpi']=400

fig = plt.figure(figsize=(3.25, 2))
# These frame numbers are specific to SC_ngait_og5
hs1 = 43
hs2 = 178
col_to = (64.-hs1)/(hs2-hs1)*100
col_hs = (109.-hs1)/(hs2-hs1)*100
to = (131.-hs1)/(hs2-hs1)*100
gaitcycle_percent = np.linspace(0,100, hs2-hs1)


m1 = ksm_mforce[hs1:hs2]-0.04
m2 = ksm_mforce[hs1:hs2]+0.04
measured_line = plt.plot(gaitcycle_percent, ksm_mforce[hs1:hs2],'-',color=(0.3,0.3,0.3) ,lw=3,zorder = -200 )
# plt.plot(gaitcycle_percent, m2, lw=0.2)
# plt.fill_between(gaitcycle_percent, m1,m2, color=(0.7,0.7,0.7), lw=0.3, edgecolor="k")


plt.fill_between(gaitcycle_percent, lsm_cforce_max[hs1:hs2],lsm_cforce_min[hs1:hs2], color='b', lw=0, alpha=0.3)
plt.fill_between(gaitcycle_percent, asm_cforce_max[hs1:hs2],asm_cforce_min[hs1:hs2], color='g', lw=0, alpha=0.5)

plt.fill_between(gaitcycle_percent, ksm_cforce_max[hs1:hs2],ksm_cforce_min[hs1:hs2],color='r', lw=0, alpha=0.5)



# plt.fill_between(gaitcycle_percent, ksm_cforce_max[hs1:hs2],ksm_cforce_min[hs1:hs2], lw=0, alpha=0.2)

plt.plot(gaitcycle_percent, asm_cforce[hs1:hs2,0], color = (0,0.6,0), lw=1)
plt.plot(gaitcycle_percent, lsm_cforce[hs1:hs2,0], color = (0,0.0,0.6), lw=1)
plt.plot(gaitcycle_percent, ksm_cforce[hs1:hs2,0], color = (0.8,0.0,0.0), lw=1)

# Create Legend
proxy_areas = [ plt.Rectangle((0, 0), 1, 1, fc=_, alpha = 0.7) for _ in ['b','g','r']]
proxy_areas.append(plt.Line2D((0,0),(1,1),lw= 2,color=(0.3,0.3,0.3) ))
modelnames = ['Linear',
              'Anatomic',
              'Kinematic',
              'Experiment']
plt.legend(proxy_areas, modelnames, prop ={'size':'x-small'},
              bbox_to_anchor=(0.95, 0.92), loc=1, borderaxespad=0.)


plt.xlim((0,100))
plt.yticks([0, 1, 2,3,4])

plt.xlabel('Gait cycle (%)')
plt.ylabel('Total knee force\n(x Body Weight)') 
plt.tight_layout()
fig.savefig('Z:\\model_output\\Figure output\\MonteCarloKneeForce.png',dpi=600)
```
